### Peter writes:

With increasing B/T e1 (and e2) go from negative to positive to no correlation. Because it affects both ellipticity components similarly, a simplified picture that can shed some light on this behavior is to consider the isotropic part of the PSF correction: When a galaxy model is larger than the actual galaxy, the PSF correction will be weaker than it needs to be, and you end up with a positive correlation (and vice versa).

This implies that disk-dominated models with low B/T may be a bit smaller than the real galaxies, maybe because the latter ones are actually pure disks but are now fit with a bulge+disk model that is more centrally concentrated.
In the intermediate regime, galaxy models would have to be too large, whereas bulge-dominated galaxies seem OK.

While all of that is conceivable, it still sounds contrived. I wonder how much of this may come from the brighter-fatter relation.

### Vinu:

I think Peter is correct about this. I plotted the histogram of the radius of these galaxies separated by B/T. Disk galaxies are the smallest, intermediate galaxies are largest and bulge dominated galaxies are in-between which is exactly Peter said. However, I noticed another issue. If you look at the distribution of RGPP_RP, which is the ratio of PSF convolved FWHM of galaxy over the FWHM of PSF, the trend is a bit different. The disk galaxies look larger with respect to the PSF size which goes exactly the opposite direction of Peter's explanation. Another thing I noticed is that the distribution of B/T looks a bit unphysical (as a galaxy morphology person) where most of the galaxies have a narrow range of B/T between 0.6 and 0.7. I am not sure how much the shape measurement groups care such things. However, if the measured shapes depend on B/T it may be considered, right?
