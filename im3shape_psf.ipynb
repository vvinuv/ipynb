```python
import os
os.chdir('/data3/scratch/DES_vinu/DES_catalogs/')
from astropy.io import fits
import numpy as np
import pylab as pl
import basics as bas

```

```python
im3 = fits.open('im3shape_v6_r_all_cols_only_galaxies_Aug8.fits')[1].data
print im3.shape
e = (im3['error_flag'] == 0) & (im3['info_flag'] == 0) & (im3['ra'] > 60) & (im3
['ra'] < 90) & (im3['dec'] > -62) & (im3['dec'] < -42)
im3 = im3[e]
```

```python
pylab inline
```

```python
bt = abs(im3['bulge_flux']) / (abs(im3['bulge_flux'])  + abs(im3['disc_flux']))
```

```python
lbt = (bt < 0.3)
ubt = (bt > 0.7)
mbt = (bt <= 0.7) & (bt >=0.3)
```

```python
bas.plot_psf(im3['MEAN_PSF_E1_SKY'], im3['E1'], -0.01, 0.03, 10, 'E1', 'k')
bas.plot_psf(im3['MEAN_PSF_E1_SKY'][lbt], im3['E1'][lbt], -0.01, 0.03, 10, 'E1-low B/T', 'r')
bas.plot_psf(im3['MEAN_PSF_E1_SKY'][ubt], im3['E1'][ubt], -0.01, 0.03, 10, 'E1-high B/T', 'g')
bas.plot_psf(im3['MEAN_PSF_E1_SKY'][mbt], im3['E1'][mbt], -0.01, 0.03, 10, 'E1-med B/T', 'b')
```
