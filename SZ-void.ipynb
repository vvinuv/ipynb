```python
def return_y_dist(file): 
    '''Returns y values within 1/2 Rv'''
    f = np.load(file)
    y = f['y']
    d_mpc = f['d_b']
    Rv = f['Rv']
    y_half_Rv = []
    for r, d, ty in zip(Rv, d_mpc, y):
        d_Rv = d / r
        y_half_Rv = np.concatenate((y_half_Rv, ty[d_Rv < 0.5]))
    return y_half_Rv
```
