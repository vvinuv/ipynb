# 5. $<e_1>$ & $<e_2>$ as function of distance to the nearest neighbour and its brighness

In following figures $m-m_n$ is the magnitude difference between the objects and the brightest of the three nearest neighbours

```python
e1_avg = []
e2_avg = []
e1_std = []
e2_std = []
e1_wavg = []
e2_wavg = []
frac = []
for td in np.arange(1, 30, 1):
    for tdm in np.arange(-2, 2, 0.2):
        con = (im3['dist_n1'] < td) & ((im3['dm1'] > tdm) | (im3['dm2'] > tdm) | (im3['dm3'] > tdm))
        fraction = con[~con].shape[0] / (con.shape[0] * 1.)
        e1_avg.append(im3['e1'][~con].mean())
        e2_avg.append(im3['e2'][~con].mean())
        e1_std.append(im3['e1'][~con].std() / np.sqrt(con[~con].shape[0]))
        e2_std.append(im3['e2'][~con].std() / np.sqrt(con[~con].shape[0]))
        e1_wavg.append(np.average(im3['e1'][~con], weights=1/im3['covmat_2_2'][~con]))
        e2_wavg.append(np.average(im3['e2'][~con], weights=1/im3['covmat_3_3'][~con]))
        frac.append(fraction)
        #print td, tdm, fraction, im3['e1'][~con].mean(), im3['e2'][~con].mean()
```

# 71. $<e_1>$ & $<e_2>$ as function of distance and magnitude of neighbours

```python
e1_avg_n = []
e2_avg_n = []
e1_avg_nn = []
e2_avg_nn = []
frac = []
m_arr = np.arange(-2, 2, 0.25)
d_arr = np.arange(5, 25, 2)
for m in m_arr:
    for d in d_arr:      
        nc = ((im3['mag_auto_r_n1'] - im3['mag_auto_r'] < m) & (im3['dist_n1'] < d))| \
        ((im3['mag_auto_r_n2'] - im3['mag_auto_r'] < m) & (im3['dist_n2'] < d))| \
        ((im3['mag_auto_r_n3'] - im3['mag_auto_r'] < m) & (im3['dist_n3'] < d))| \
        ((im3['mag_auto_r_n4'] - im3['mag_auto_r'] < m) & (im3['dist_n4'] < d))| \
        ((im3['mag_auto_r_n5'] - im3['mag_auto_r'] < m) & (im3['dist_n5'] < d))
        #print m, d
        e1_avg_n.append(im3['e1'][nc].mean())
        e2_avg_n.append(im3['e2'][nc].mean())
        e1_avg_nn.append(im3['e1'][~nc].mean())
        e2_avg_nn.append(im3['e2'][~nc].mean())
        frac.append(nc[~nc].shape[0] * 1. / nc.shape[0])
e1_avg_n = np.array(e1_avg_n)
e2_avg_n = np.array(e2_avg_n)
e1_avg_nn = np.array(e1_avg_nn)
e2_avg_nn = np.array(e2_avg_nn)
frac = np.array(frac)
```
