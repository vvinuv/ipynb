```python
print np.unique(im3['im3shape_r_stamp_size'])
```

```python
for ss in [32, 48, 64, 96]:
    con = (im3['im3shape_r_stamp_size'] == ss)
    pl.figure(99)
    xxx = pl.hist(im3['mask_fraction'][con], 20, histtype='step', label='SS=%d'%ss)
pl.legend(loc=0)
```

```python
for ss in [32, 48, 64, 96]:
    A = ss * ss
    print A, A*0.3
```

```python
con = (im3['im3shape_r_stamp_size'] == 32) & (im3['mask_fraction'] > 0.25)
print con[con].shape[0] * 1.0 / con.shape[0]
```

```python
e1_avg = []
e2_avg = []
e1_std = []
e2_std = []
N = []
mf_arr = np.arange(0.1, 0.5, 0.05)
mff3 = []
mff4 = []
for mf3 in mf_arr:
    for mf4 in mf_arr:
        con = ((im3['im3shape_r_stamp_size'] == 32) & (im3['mask_fraction'] > mf3)) | \
            ((im3['im3shape_r_stamp_size'] == 48) & (im3['mask_fraction'] > mf4))
        e1_avg.append(im3['e1'][~con].mean())
        e2_avg.append(im3['e2'][~con].mean())
        e1_std.append(im3['e1'][~con].std())
        e2_std.append(im3['e2'][~con].std())
        N.append(con[~con].shape[0])
        mff3.append(mf3)
        mff4.append(mf4)
e1_avg = np.array(e1_avg)
e2_avg = np.array(e2_avg)
e1_std = np.array(e1_std)
e2_std = np.array(e2_std)
N = np.array(N)
e1_std = e1_std / np.sqrt(N)
e2_std = e2_std / np.sqrt(N)
        #print con[~con].shape[0] * 1.0 / con.shape[0], im3['e1'][~con].mean(), im3['e2'][~con].mean()
```

```python
for t, u,v,w,x,y,z in zip(e1_avg, e2_avg, e1_std, e2_std, N*1./im3.shape[0], mff3, mff4):
    if abs(t) < 6e-4 and abs(u) < 6e-4:
        print t, u, v, w, x, y, z
```

```python
pl.figure(figsize=(24, 5))
pl.subplot(131)
pl.title(r'$\langle e_1 \rangle$')
pl.imshow(e1_avg.reshape(mf_arr.shape[0], mf_arr.shape[0]), interpolation='none', origin='lower')
pl.xticks(range(mf_arr.shape[0]), ['%.1f'%i for i in mf_arr])
pl.yticks(range(mf_arr.shape[0]), ['%.1f'%i for i in mf_arr])
pl.xlabel('Mask fraction (SS=32)')
pl.ylabel('Mask fraction (SS=48)')
pl.colorbar(shrink=0.9)
pl.subplot(132)
pl.title(r'$\sigma_{e_1}$')
pl.imshow(e1_std.reshape(mf_arr.shape[0], mf_arr.shape[0]), interpolation='none', origin='lower')
pl.xticks(range(mf_arr.shape[0]), ['%.1f'%i for i in mf_arr])
pl.yticks(range(mf_arr.shape[0]), ['%.1f'%i for i in mf_arr])
pl.xlabel('Mask fraction (SS=32)')
#pl.ylabel('Mask fraction (SS=48)')
pl.colorbar(shrink=0.9)
pl.subplot(133)
pl.title(r'$\langle e_1 \rangle / \sigma_{e_1}$')
pl.imshow((e1_avg/e1_std).reshape(mf_arr.shape[0], mf_arr.shape[0]), interpolation='none', origin='lower')
pl.xticks(range(mf_arr.shape[0]), ['%.1f'%i for i in mf_arr])
pl.yticks(range(mf_arr.shape[0]), ['%.1f'%i for i in mf_arr])
pl.xlabel('Mask fraction (SS=32)')
#pl.ylabel('Mask fraction (SS=48)')
pl.colorbar(shrink=0.9)
```
